In the radioactive chain Bi210  Po210  Pb206 (bismuth-210 to polo-
nium-210 to lead-206), the decay rate of Bi210, a, is 0.0137/da and the

decay rate of Po210, b, is 0.0051/da. Assuming the initial mass of Bi210 is

108 g and using your model from Part a, find, approximately, the maxi-
mum mass of Po210 and when the maximum occurs.

```python
a = 0.00000117
b = 0.181
A_0 = 1

A = A_0 #Initial Value of A
B = 0
C = 0
delT = 0.1

time = []
substanceA = []
substanceB = []
substanceC = []

B_max = 0;
B_max_time = 0;

for i in np.arange(0, 700, delT):

  time = np.append(time, i)
  A_rem = a*A*delT
  substanceA = np.append(substanceA, A)
  A = (A - A_rem)
  substanceB = np.append(substanceB,B)
  B_rem = B*b*delT
  C = C + B_rem
  B = B + A_rem - B_rem
  if (B_max < B):
    B_max = B
    B_max_time = i

  substanceC = np.append(substanceC, C)


# plot of the amount of substanceA, substanceB and substanceC
# print(B_max)
# print(max(substanceB))
# plt.plot(time, substanceA, 'b-.', label='Bi210')
# plt.plot(time, substanceB, 'r', label='Po210')
plt.plot(time, substanceB/substanceA, 'r', label='Po210')
# plt.plot(time, substanceC, 'g-.', label='Pb206')
plt.vlines(x = B_max_time, ymin = 0, ymax = B_max, color = 'm', label='Max. Value of Po210')

plt.xlabel('Time (days)')
plt.ylabel('Amount (g)')
plt.grid(True)
plt.scatter(B_max_time, B_max)
plt.text(B_max_time, B_max, B_max)
# plt.text(B_max_time, B_max, B_max_time)
plt.legend()
```
